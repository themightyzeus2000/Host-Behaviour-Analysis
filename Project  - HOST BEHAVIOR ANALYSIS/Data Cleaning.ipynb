```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Hp/Downloads/availability_final.csv")
```

```python
# Checking the null values in the file
df.isnull().sum()
```

```python
# Replacing the null values in the file with the average values of the particular column
df['price'].mean()
df['price'].fillna(value=int(df['price'].mean()), inplace=True)
```

```python
# Confirming the replacement of the null values 
df.isnull().sum()
```

```python
# Replacing for other columns
df['adjusted_price'].fillna(value=int(df['adjusted_price'].mean()), inplace=True)
df['minimum_nights'].fillna(value=int(df['minimum_nights'].mean()), inplace=True)
df['maximum_nights'].fillna(value=int(df['maximum_nights'].mean()), inplace=True)
```

```python
df.isnull().sum()
```

```python
# Extracting the Data
df.to_csv("C:/Users/Hp/Desktop/availability_final.csv", index = False)
```
